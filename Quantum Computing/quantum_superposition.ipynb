# Quantum Superposition

## Cirq

```python
import matplotlib.pyplot as plt 
import numpy as np

import cirq
```

```python
q0 = cirq.NamedQubit("q0")
```

```python
circuit = cirq.Circuit(
    cirq.measure(q0, key='measure_all')
)
```

```python
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions = 1024)
```

```python
cirq.plot_state_histogram(result, plt.subplot())
```

```python
circuit = cirq.Circuit(
    cirq.H(q0),
    cirq.measure(q0, key='measure_all')
)
```

```python
print(circuit)
```

```python
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions = 1024)
```

```python
cirq.plot_state_histogram(result, plt.subplot())
```

## Qiskit

```python
import qiskit
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)
qc.measure_all()
```

```python
from qiskit import IBMQ, assemble, Aer, BasicAer, transpile, execute
from qiskit.visualization import plot_histogram, plot_bloch_vector

backend = Aer.get_backend('aer_simulator') # Tell Qiskit how to simulate our circuit 

results = execute(qc,backend).result().get_counts()
plot_histogram(results)
```

```python
# Apply H-gate to the first:
qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()
qc.draw('mpl')
```

```python
from qiskit import IBMQ, assemble, Aer, BasicAer, transpile, execute
from qiskit.visualization import plot_histogram, plot_bloch_vector

backend = Aer.get_backend('aer_simulator') # Tell Qiskit how to simulate our circuit 

results = execute(qc,backend).result().get_counts()
plot_histogram(results)
```
