# Kelly Formula

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def toss(p):
    return 'H' if random.random() < p else 'T'
```

```python
[toss(0.5) for i in range(10)]
```

```python
N = 10
p = 0.5
B = 1
balance = 100.
balances = []

f = p - (1-p)/B

for i in range(N): 
        balances.append(balance)
        b = toss(p)
        if b == 'H':
            balance += balance * f * B
        else:
            balance -= balance * f

plt.plot(balances)
plt.ylabel('money')
plt.show()
```

```python
N = 10000
p = 0.5
B = 2
balances = []

f_range = np.arange(0,1,0.05)
for f in f_range:
    balance = 100
    for i in range(N): 
        b = toss(p)
        if b == 'H':
            balance += balance * f * B
        else:
            balance -= balance * f
    balances.append(balance)
    
plt.plot(f_range,balances)
plt.ylabel('money')
plt.show()
```
