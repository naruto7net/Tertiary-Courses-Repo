# Matplotlib Subplot Layout

## Import Data

```python
import tensorflow as tf
from tensorflow import keras

cifar10 = keras.datasets.cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
for i in range(6):
  print(f'Label: {y_train[i]}')
  plt.imshow(x_train[i], cmap='gray')
  plt.show()
```

## Method 1

```python
import matplotlib.pyplot as plt
import numpy as np

ncols,nrows = 3, 2
fig, ax = plt.subplots(nrows, ncols,figsize=(7,5), constrained_layout=True)

for i in range(nrows):
    for j in range(ncols):
        ax[i,j].imshow(x_train[i*ncols+j])
```

## Method 2

```python
fig = plt.figure(figsize=(7, 5), constrained_layout=True)
spec = fig.add_gridspec(nrows, ncols)

for i in range(nrows):
    for j in range(ncols):
        ax = fig.add_subplot(spec[i, j])
        ax.imshow(x_train[i*ncols+j])
```

```python
fig = plt.figure(figsize=(7, 5), constrained_layout=True)
spec = fig.add_gridspec(2, 3)

ax0 = fig.add_subplot(spec[0, :])
ax10 = fig.add_subplot(spec[1, 0])
ax11 = fig.add_subplot(spec[1, 1:3])
```
