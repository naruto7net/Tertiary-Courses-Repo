# Augmented Dickey Fuller (ADF) Test for Stationary

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
import yfinance as yf
```

```python
import yfinance as yf
start_date = '2020-06-01'

data = yf.download('Z74.SI', start = start_date)
data['Adj Close'].plot(figsize=(12,8))
```

```python
# ADF Test

adf_result= adfuller(data['Adj Close'])
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
```

## Returns

```python
data['returns'] = data['Adj Close'].pct_change()
data.dropna(inplace=True)
data['returns'].plot(figsize=(12,8))
```

```python
# ADF Test

adf_result= adfuller(data['returns'])
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
```
