
# AutoRegessive (AR) Model 


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.tsa.api as smtsa
```

```python
# Function to plot signal, ACF and PACF
def plotds(xt, nlag=30, fig_size=(12, 10)):
    if not isinstance(xt, pd.Series):
         xt = pd.Series(xt)
    plt.figure(figsize=fig_size)
    layout = (2, 2)
    
    # Assign axes
    ax_xt = plt.subplot2grid(layout, (0, 0), colspan=2)
    ax_acf= plt.subplot2grid(layout, (1, 0))
    ax_pacf = plt.subplot2grid(layout, (1, 1))
    
    # Plot graphs
    xt.plot(ax=ax_xt)
    ax_xt.set_title('Time Series')
    plot_acf(xt, lags=50, ax=ax_acf)
    plot_pacf(xt, lags=50, ax=ax_pacf)
    plt.tight_layout()
    return None
    
```

## AR(1) Model

```python
# Simulate AR(1) data

ar = np.array([1, -0.5])
ma = np.array([1])
data = smtsa.arma_generate_sample(ar,ma,1000) 
plotds(data)
```

```python
# Build AR(1) model
model = ARIMA(data, order=(1, 0, 0))

## Fitting the model, disp=0 is to switch off verbose display
trained_model = model.fit()

## Printing a summary of the model
trained_model.summary()
```

## AR Model on Stock Data

```python
import yfinance as yf

data = yf.download('MSFT', start = '2011-1-1', end = '2021-1-1')

plotds(data['Adj Close'].values)
```

```python
model = ARIMA(data['Adj Close'].values, order=(1, 0, 0))
trained_model = model.fit()
trained_model.summary()
```
