# Value Based Q-Learning 

## Import Python Libraries

```python
import numpy as np
import gym
import time

lr = 0.1
gamma = 0.8
epsilon = 0.1
episodes = 50000

env = gym.make('Taxi-v3')
```

## Initialize Q Table

```python
Q = np.zeros((env.observation_space.n, env.action_space.n))
print(Q)
```

## Before Q-learning

```python
s = env.reset()
done = False
step_count = 0
total_reward = 0

while not done:
    env.render()
    a = np.argmax(Q[s,:])
    s_, r, done, _ = env.step(a)
    s = s_
    step_count += 1
    total_reward += r
    time.sleep(0.1)

print("Total steps: ",step_count)
print("Total rewards: ",total_reward)
```

## Q-Learning

```python

for i in range(episodes):
    if i%2000 == 0: print("Episode {}/{}".format(i + 1, episodes))
    s = env.reset()
    done = False

    while not done:
        if np.random.random() < epsilon:
            a = env.action_space.sample()
        else:
            a = np.argmax(Q[s,:])
        s_, r, done, _ = env.step(a)
        Q[s,a] += lr*(r+gamma*np.max(Q[s_,:]) - Q[s,a])
        s = s_
```

```python
# Print Final Q Table
print(Q)
```

## After Q-learning

```python
s = env.reset()
done = False
step_count = 0
total_reward = 0

while not done:
    env.render()
    a = np.argmax(Q[s,:])
    s_, r, done, _ = env.step(a)
    s = s_
    step_count += 11
    total_reward += r
    time.sleep(0.1)

print("Total steps: ",step_count)
print("Total rewards: ",total_reward)
```
