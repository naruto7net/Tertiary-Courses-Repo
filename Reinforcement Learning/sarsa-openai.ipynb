# Value-Based SARSA Reinforcement Learning

## Import Python Libraries

```python
ﬁﬁimport numpy as np
import gym
import time

#env = gym.make('FrozenLake-v0')
env = gym.make("Taxi-v3").env
```

```python
## Initialize Q Table
Q = np.zeros((env.observation_space.n, env.action_space.n))
print(Q)
```

```python
## Before SARSA Learning

s = env.reset()
done = False
step_count = 0
total_reward = 0

while not done:
    env.render()
    a = np.argmax(Q[s,:])
    s_, r, done, _ = env.step(a)
    s = s_
    step_count += 1
    total_reward += r
    time.sleep(0.1)
    if step_count > 200: break

print("Total steps: ",step_count)
print("Total rewards: ",total_reward)
```

```python
## Action Policy

def choose_action(s, Q):
    if np.random.random() < epsilon:
        a = np.random.randint(0, env.action_space.n)
    else:
        a = np.argmax(Q[s,:])
    return a
```

```python
## SARSA

lr = 0.05
gamma = 0.9
epsilon = 0.1
episodes = 50000

for i in range(episodes):
    if i%2000 == 0: print("Episode {}/{}".format(i + 1, episodes))
    s = env.reset()
    done = False
        
    a = choose_action(s, Q)
    while not done:
        s_, r, done, _ = env.step(a)
        a_ = choose_action(s_, Q)
        Q[s,a] = Q[s,a] + lr*(r+gamma*(Q[s_,a_]) - Q[s,a])
        s = s_
        a = a_
```

```python
# Print Final Q Table
print(Q)
```

```python
## After SARSA Learning

s = env.reset()
done = False
step_count = 0
total_reward = 0

while not done:
    env.render()
    a = np.argmax(Q[s,:])
    s_, r, done, _ = env.step(a)
    s = s_
    step_count += 1
    total_reward += r
    time.sleep(0.1)
    if step_count > 200: break

print("Total steps: ",step_count)
print("Total rewards: ",total_reward)
```
