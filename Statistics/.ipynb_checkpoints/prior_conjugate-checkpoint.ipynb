```python
import numpy as np
import scipy.stats
import random
import matplotlib.pyplot as plt
```

```python
def coin_flip(p):
    return 1 if random.random() < p else 0
```

```python
def plot_prior_beta(alpha, beta):
    r=np.arange(0.,1.01,.01)
    fig, ax = plt.subplots(figsize=(5,5))
    plt.title(f"Prior Beta Distribution({alpha},{beta})")
    ax.plot(r, scipy.stats.beta.pdf(r, alpha, beta), 'r-', lw=2, alpha=0.6, label='beta pdf')
    ax.set_xlabel("$\\theta$")
    ax.set_ylabel("prior probability density")  
    plt.show()
```

```python
def monte_carlo(n):
  head = 0
  for i in range(n):
    flip_result = coin_flip(p)
    head = head + flip_result
  return head
```

```python
p=0.9
n = 5
alpha,beta = 1,1
for i in range(100):
    
    if i%10==0: plot_prior_beta(alpha, beta)
    
    k = monte_carlo(n)
    alpha,beta = alpha+k,beta+n-k
    
```
